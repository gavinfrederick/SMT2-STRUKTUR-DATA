Binary Tree

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Binary Search Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)
        return root

    def inorder(self, root):
        if root is not None:
            self.inorder(root.left)
            print(root.data, end=" ")
            self.inorder(root.right)

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]

for value in data_list:
    root = bst.insert(root, value)

print("Output BST (Inorder Traversal):")
bst.inorder(root)
```

Travelsar pada BST

```python
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

Pencarian Nilai dalam BST

```python
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)

# Uji pencarian
key = 60
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```

Tugas 1 (Program Struktur Binary Tree Manual)

```python
# Struktur Tree berdasarkan identitas saya
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def inorder_traversal(node):
    if node is not None:
        inorder_traversal(node.left)
        print(node.data, end=" ")
        inorder_traversal(node.right)

# === IDENTITAS ===
nama_lengkap = "Gavin Frederick Purnomo"
dua_digit_npm = 73
tanggal_lahir = 15
bulan_lahir = 8

# === PENGOLAHAN DATA ===
nama_tanpa_spasi = nama_lengkap.replace(" ", "")
jumlah_huruf = len(nama_tanpa_spasi)

# Menghitung jumlah huruf vokal
huruf_vokal = "aiueoAIUEO"
jumlah_vokal = sum(1 for huruf in nama_tanpa_spasi if huruf in huruf_vokal)

# === MEMBANGUN BINARY TREE SECARA MANUAL ===
root = Node(jumlah_huruf)
# Anak kiri root: jumlah huruf vokal
root.left = Node(jumlah_vokal)
# Anak kanan root: dua digit terakhir NPM
root.right = Node(dua_digit_npm)
# Anak kiri dari node kiri: tanggal (DD)
root.left.left = Node(tanggal_lahir)
# Anak kanan dari node kanan: bulan (MM)
root.right.right = Node(bulan_lahir)

print("=== Rincian Data Tree ===")
print(f"Jumlah Huruf nama     : {jumlah_huruf}")
print(f"Jumlah vokal          : {jumlah_vokal}")
print(f"2 digit NPM           : {dua_digit_npm}")
print(f"Tanggal               : {tanggal_lahir}")
print(f"Bulan                 : {bulan_lahir}\n")

print("Output Inorder Traversal:")
inorder_traversal(root)
print()
```

Tugas 2 (Program Binary Search Tree)

```python
# Class untuk merepresentasikan Node pada BST
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# Fungsi untuk menyisipkan node baru ke dalam BST secara otomatis
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val == key:
            return root
        elif key < root.val:
            root.left = insert(root.left, key)
        else:
            root.right = insert(root.right, key)
    return root

# Fungsi untuk Inorder Traversal (Kiri -> Root -> Kanan)
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

# Fungsi untuk Preorder Traversal (Root -> Kiri -> Kanan)
def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

# Fungsi untuk Postorder Traversal (Kiri -> Kanan -> Root)
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

# Fungsi untuk mencari nilai di dalam BST
def search(root, key):
    if root is None or root.val == key:
        return root
    if key < root.val:
        return search(root.left, key)
    return search(root.right, key)

# BAGIAN PENGISIAN DATA IDENTITAS

tiga_digit_npm = 263
dua_digit_npm = 73
tanggal_lahir = 15
bulan_lahir = 8
angka_bebas_1 = 3
angka_bebas_2 = 29

# Mengambil nilai ASCII dari huruf 'G' dan 'a' (dari nama "Gavin")
ascii_huruf_1 = ord('G')
ascii_huruf_2 = ord('a')

# Membuat daftar berisi 7 angka sesuai ketentuan
data_numerik = [
    tiga_digit_npm,
    tanggal_lahir,
    bulan_lahir,
    ascii_huruf_1,
    ascii_huruf_2,
    angka_bebas_1,
    angka_bebas_2
]

# PEMBENTUKAN BST & EKSEKUSI
root = None

# Menyisipkan semua angka dari data_numerik ke dalam BST
for angka in data_numerik:
    root = insert(root, angka)

print(f"Data Numerik (sebelum BST) : {data_numerik}\n")

# Menampilkan hasil dari ketiga jenis traversal
print("--- Hasil Traversal ---")
print("Inorder   :", end=" ")
inorder(root)
print("\nPreorder  :", end=" ")
preorder(root)
print("\nPostorder :", end=" ")
postorder(root)
print("\n")

# Menjalankan fitur pencarian
print("--- Hasil Pencarian ---")

# 1. Mencari angka dua digit terakhir NPM
if search(root, dua_digit_npm):
    print(f"Pencarian {dua_digit_npm} (2 Digit NPM): DITEMUKAN dalam BST.")
else:
    print(f"Pencarian {dua_digit_npm} (2 Digit NPM): TIDAK DITEMUKAN dalam BST.")

# 2. Mencari angka yang tidak ada di dalam list
angka_tidak_ada = 999
if search(root, angka_tidak_ada):
    print(f"Pencarian {angka_tidak_ada} (Di luar list): DITEMUKAN dalam BST.")
else:
    print(f"Pencarian {angka_tidak_ada} (Di luar list): TIDAK DITEMUKAN dalam BST.")
```
