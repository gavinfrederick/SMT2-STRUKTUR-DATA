Bubble Sort

```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n - 1):
    for j in range(n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
  return arr

data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort(data.copy()) # Menggunakan copy agar data asli tidak berubah
print("Data sebelum sorting:", data)
print("Data setelah Bubble Sort:", sorted_data)
```

Selection Sort

```python
def selection_sort(arr):
  n = len(arr)
  for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
  return arr


data = [64, 25, 12, 22, 11]
sorted_data = selection_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Selection Sort:", sorted_data)
```

Insertion Sort

```python
def insertion_sort(arr):
  n = len(arr)
  for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
      arr[j + 1] = key
  return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Insertion Sort:", sorted_data)
```

Merge Sort

```python
def merge_sort(arr):
  if len(arr) > 1:
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i = j = k = 0

    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] = right_half[j]
        j += 1
        k += 1

    while i < len(left_half):
      arr[k] = left_half[i]
      i += 1
      k += 1

    while j < len(right_half):
      arr[k] = right_half[j]
      j += 1
      k += 1
  return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Merge Sort:", sorted_data)
```

Quick Sort

```python
def quick_sort(arr):
  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Quick Sort:", sorted_data)
```

Tugas

```python
def quick_sort_desc(data, kriteria):
    """
    Mengurutkan daftar dictionary dari nilai tertinggi ke terendah (Descending).
    Parameter 'kriteria' memungkinkan pengurutan fleksibel berdasarkan key apapun (misal: 'rating' atau 'jumlah_tayang').
    """
    # Base case: jika list kosong atau hanya berisi 1 elemen, list sudah terurut
    if len(data) <= 1:
        return data

    # Menentukan pivot (mengambil elemen tengah)
    pivot = data[len(data) // 2]

    # Partisi array menjadi tiga bagian berdasarkan kriteria
    left = [x for x in data if x[kriteria] > pivot[kriteria]]
    middle = [x for x in data if x[kriteria] == pivot[kriteria]]
    right = [x for x in data if x[kriteria] < pivot[kriteria]]

    # Rekursi untuk bagian kiri dan kanan, lalu gabungkan
    return quick_sort_desc(left, kriteria) + middle + quick_sort_desc(right, kriteria)


# --- SKENARIO PENGGUNAAN ---

# Dummy data: Daftar film beserta atributnya
daftar_film = [
    {"judul": "The Dark Knight", "genre": "Action", "rating": 9.0, "jumlah_tayang": 3200000},
    {"judul": "Inception", "genre": "Sci-Fi", "rating": 8.8, "jumlah_tayang": 2500000},
    {"judul": "Parasite", "genre": "Thriller", "rating": 8.5, "jumlah_tayang": 1800000},
    {"judul": "Interstellar", "genre": "Sci-Fi", "rating": 8.6, "jumlah_tayang": 4100000},
    {"judul": "Avengers: Endgame", "genre": "Action", "rating": 8.4, "jumlah_tayang": 5500000}
]

print("=== DAFTAR FILM SEBELUM DIURUTKAN ===")
for film in daftar_film:
    print(f"- {film['judul']} | Rating: {film['rating']} | Tayang: {film['jumlah_tayang']}")

# 1. Mengurutkan berdasarkan Rating Tertinggi
film_berdasarkan_rating = quick_sort_desc(daftar_film, 'rating')

print("\n=== REKOMENDASI BERDASARKAN RATING TERTINGGI ===")
for i, film in enumerate(film_berdasarkan_rating, start=1):
    print(f"{i}. {film['judul']} (Rating: {film['rating']})")

# 2. Mengurutkan berdasarkan Jumlah Tayang Terbanyak
film_berdasarkan_tayang = quick_sort_desc(daftar_film, 'jumlah_tayang')

print("\n=== REKOMENDASI BERDASARKAN JUMLAH TAYANG TERBANYAK ===")
for i, film in enumerate(film_berdasarkan_tayang, start=1):
    print(f"{i}. {film['judul']} (Tayang: {film['jumlah_tayang']} kali)")
```
