Hapus Node

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add node in the begining
    def AddBegining(self, newdata):
        NewNode = Node(newdata)
        # Update the new nodes next val to existing no
        NewNode.nextval = self.headval
        self.headval = NewNode

    # Function to remove node by key
    def RemoveNode(self, Removekey):
        HeadVal = self.headval

        # Jika node yang ingin dihapus ada di head (paling depan)
        if (HeadVal is not None):
            if (HeadVal.dataval == Removekey):
                self.headval = HeadVal.nextval
                HeadVal = None
                return

        # Mencari node yang ingin dihapus, simpan node sebelumnya (prev)
        while (HeadVal is not None):
            if HeadVal.dataval == Removekey:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval

        # Jika key tidak ditemukan
        if (HeadVal == None):
            return

        # Putuskan node dari linked list
        prev.nextval = HeadVal.nextval
        HeadVal = None

# create linked list named Li
Li = LinkedList()

# it add node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Febuari")

print("Sebelum dilakukan remove:")
Li.listprint()
print("-" * 30)

# remove node by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret':")
Li.listprint()
```

Hapus Node Diakhir

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add newnode in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return
        last = self.headval
        while(last.nextval):
            last = last.nextval
        last.nextval = NewNode

    # Function to remove last node
    def RemoveEnd(self):
        last = self.headval
        while (last is not None):
            if last.nextval == None:
                break
            prev = last
            last = last.nextval

        if (last == None):
            return

        prev.nextval = last.nextval
        last = None

# create linked list named Li
Li = LinkedList()

# add node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")

print("Sebelum dilakukan remove:")
Li.listprint()
print(" =======")

# remove last node
Li.RemoveEnd()

print("Setelah dilakukan remove node terakhir:")
Li.listprint()
```

1. Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan
tengah dan hapus dengan kata kunci !

```python
# --- SOAL 1: DOUBLY LINKED LIST ---

class DLLNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Fungsi bantu untuk menambahkan node di akhir (untuk membuat 3 node awal)
    def append(self, data):
        new_node = DLLNode(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # Fungsi penyisipan tengah (disisipkan setelah node target tertentu)
    def insert_tengah(self, target_data, new_data):
        curr = self.head
        while curr:
            if curr.data == target_data:
                new_node = DLLNode(new_data)
                new_node.next = curr.next
                curr.next = new_node
                new_node.prev = curr
                if new_node.next:
                    new_node.next.prev = new_node
                print(f"Berhasil menyisipkan '{new_data}' setelah '{target_data}'.")
                return
            curr = curr.next
        print(f"Node dengan data '{target_data}' tidak ditemukan.")

    # Fungsi hapus dengan kata kunci
    def delete_by_key(self, key):
        curr = self.head
        while curr:
            if curr.data == key:
                # Jika node yang dihapus adalah head
                if curr.prev is None:
                    self.head = curr.next
                else:
                    curr.prev.next = curr.next

                # Jika node yang dihapus bukan node terakhir
                if curr.next is not None:
                    curr.next.prev = curr.prev

                print(f"Berhasil menghapus node dengan kunci '{key}'.")
                return
            curr = curr.next
        print(f"Kunci '{key}' tidak ditemukan.")

    def display(self):
        curr = self.head
        elements = []
        while curr:
            elements.append(str(curr.data))
            curr = curr.next
        print(" <-> ".join(elements))


# --- EKSEKUSI SOAL 1 ---
print("=== DEMO DOUBLY LINKED LIST ===")
dll = DoublyLinkedList()

# 1. Buat dengan 3 node didalamnya
dll.append("A")
dll.append("B")
dll.append("C")
print("Kondisi awal (3 node):")
dll.display()

# 2. Lakukan penyisipan tengah (Misal: sisipkan "X" setelah "B")
dll.insert_tengah("B", "X")
dll.display()

# 3. Hapus dengan kata kunci (Misal: hapus "A")
dll.delete_by_key("A")
dll.display()
print("\n")
```

2. Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan
akhir dan hapus akhir!

```python
# --- SOAL 2: CIRCULAR LINKED LIST ---

class CLLNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Fungsi penyisipan akhir (juga digunakan untuk inisialisasi node awal)
    def insert_akhir(self, data):
        new_node = CLLNode(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            print(f"Berhasil menyisipkan '{data}' di akhir.")
            return

        curr = self.head
        while curr.next != self.head:
            curr = curr.next

        curr.next = new_node
        new_node.next = self.head
        print(f"Berhasil menyisipkan '{data}' di akhir.")

    # Fungsi hapus akhir
    def delete_akhir(self):
        if not self.head:
            print("List kosong, tidak ada yang bisa dihapus.")
            return

        # Jika hanya ada 1 node
        if self.head.next == self.head:
            hapus = self.head.data
            self.head = None
            print(f"Berhasil menghapus '{hapus}' di akhir.")
            return

        curr = self.head
        prev = None
        while curr.next != self.head:
            prev = curr
            curr = curr.next

        hapus = curr.data
        prev.next = self.head # Putuskan sambungan ke node terakhir, sambungkan prev ke head
        print(f"Berhasil menghapus '{hapus}' di akhir.")

    def display(self):
        if not self.head:
            print("List kosong")
            return
        curr = self.head
        elements = []
        while True:
            elements.append(str(curr.data))
            curr = curr.next
            if curr == self.head:
                break
        print(" -> ".join(elements) + " -> (kembali ke head: " + str(self.head.data) + ")")


# --- EKSEKUSI SOAL 2 ---
print("=== DEMO CIRCULAR LINKED LIST ===")
cll = CircularLinkedList()

# 1. Buat dengan 3 node didalamnya
cll.insert_akhir(10)
cll.insert_akhir(20)
cll.insert_akhir(30)
print("Kondisi awal (3 node):")
cll.display()

# 2. Lakukan penyisipan akhir
cll.insert_akhir(40)
print("Setelah penyisipan akhir:")
cll.display()

# 3. Hapus akhir
cll.delete_akhir()
print("Setelah hapus akhir:")
cll.display()
```
