Representasi Graf dengan Adjacency List


```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}


# Menampilkan graf
def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")


print_graph(graph)
```

Traversal Graf menggunakan BFS

```python
from collections import deque


def bfs(graph, start):
    visited = set()
    queue = deque([start])


    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])


# Contoh pemanggilan
bfs(graph, 'A')
```

Traversal Graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()


    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)


# Contoh pemanggilan
dfs(graph, 'A')
```

```python
def count_neighbors(graph):
    result = {}
    for node in graph:
        result[node] = len(graph[node])
    return result

count_neighbors(graph)
```

3. Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan, bukan hanya mencetak.

```python
from collections import deque

def dfs_print(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs_print(graph, neighbor, visited)

def bfs_return_list(graph, start):
    visited = set()
    queue = deque([start])
    order = []

    visited.add(start)

    while queue:
        vertex = queue.popleft()
        order.append(vertex)

        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return order

print("\n\n=== 2 & 3. Hasil Traversal ===")
print("DFS (Cetak Langsung) mulai dari 'A': ", end="")
dfs_print(my_graph, 'A')

print("\nBFS (Return List) mulai dari 'A':", bfs_return_list(my_graph, 'A'))
```

4. Buat fungsi find_path(graph, start, end) yang mengembalikan satu jalur dari start ke end jika ada, menggunakan DFS.

```python
def find_path(graph, start, end, path=None):
    if path is None:
        path = []

    path = path + [start]

    if start == end:
        return path

    if start not in graph:
        return None

    for node in graph[start]:
        if node not in path:
            new_path = find_path(graph, node, end, path)
            if new_path:
                return new_path

    return None

print("\n=== 4. Mencari Jalur ===")
print("Jalur dari 'A' ke 'E':", find_path(my_graph, 'A', 'E'))
print("Jalur dari 'D' ke 'A':", find_path(my_graph, 'D', 'A'))
```

5. ambahkan fungsi is_connected(graph) untuk mengecek apakah semua simpul dalam graf saling terhubung (gunakan kombinasi DFS + logika).


```python
def is_connected(graph):
    if not graph:
        return True

    start_node = list(graph.keys())[0]
    visited = set()

    def dfs_helper(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_helper(neighbor)

    dfs_helper(start_node)

    return len(visited) == len(graph)

print("\n=== 5. Cek Keterhubungan ===")
print("Apakah 'my_graph' terhubung?", is_connected(my_graph))

disconnected_graph = {
    'A': ['B'],
    'B': ['A'],
    'C': ['D'],
    'D': ['C']
}
print("Apakah 'disconnected_graph' terhubung?", is_connected(disconnected_graph))
```
