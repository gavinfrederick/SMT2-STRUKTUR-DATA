Stack dengan list

```python
# program python untuk implementasi stack memakai list
stack = []

# menambahkan stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Inisial stack')
print(stack)

# fungsi pop() untuk menghapus dari stack dalam LIFO

print("\nElemen yang di pop dari stack")
print(stack.pop())
print(stack.pop())
print(stack.pop())
```

Stack dengan collections.deque

```python
# program python untuk implementasi stack memakai collections.deque
from collections import deque

stack = deque()

# menambahkan stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Inisial stack')
print(stack)

# fungsi pop() untuk menghapus dari stack dalam LIFO
print("\nElemen yang di pop dari stack")
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack setelah di pop')
print(stack)
```

Stack dengan queue.LIFOQueue

```python
# program python untuk implementasi stack memakai modul queue
from queue import LifoQueue

# Inisialisasi stack
stack = LifoQueue(maxsize=5)

# qsize menampilkan nomor dari elemen didalam stack
print(stack.qsize())

# put() untuk menambahkan elemen didalam stack
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print("Full: ", stack.full())
print("Size: ", stack.qsize())

# menghapus elemen didalam stack menggunakan fungsi get()
print("\nElemen yang di pop dari stack")
print(stack.get())
print(stack.get())
print(stack.get())

print("\nEmpy: ", stack.empty())
```

Stack dengan single linked list

```python

class Node: # Diperbaiki: menggunakan huruf kecil 'c'
  def __init__(self, value):
    self.value = value
    self.next = None

class stack:
  # Inisialisasi stack, menggunakan dummy node
  def __init__(self):
    self.head = Node("head")
    self.size = 0

  # representasi string dari stack
  def __str__(self):
    cur = self.head.next
    out = ""
    while cur:
      out += str(cur.value) + "->"
      cur = cur.next
    return out[:-2] if out else "" # Tambahan keamanan jika stack benar-benar kosong

  # mendapatkan ukuran stack saat ini
  def getSize(self):
    return self.size

  # cek jika stack kosong
  def isEmpty(self):
    return self.size == 0

  # menambahkan item teratas di stack
  def peek(self):
    # memeriksa sanitasi untuk melihat stack yang kosong
    if self.isEmpty():
      raise Exception("Mengintip dari stack yang kosong")
    return self.head.next.value

  # menambahkan value ke stack
  def push(self, value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  # menghapus value dari stack dan mengembalikannya
  def pop(self):
    if self.isEmpty():
      raise Exception("Pop dari stack kosong")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

# menjalankan program
if __name__ == "__main__": # Diperbaiki: menggunakan "__main__"
  my_stack = stack() # Diperbaiki: mengubah nama variabel agar tidak bentrok dengan nama class

  for i in range(1, 11):
    my_stack.push(i)
  print(f"Stack: {my_stack}")

  for _ in range(1, 6):
    remove = my_stack.pop()
    print(f"Pop: {remove}")
  print(f"Stack: {my_stack}")
```
